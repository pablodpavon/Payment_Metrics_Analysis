# Analysis / customers

## Completed payments

```python
payments_df
```

We will analyze the total transaction amounts for each client and country, with a focus on completed transactions.

```python
payments_df.info()
```

```python
completed_payments_df = payments_df[payments_df['status_group'] == 'completed']
aggregated_df = completed_payments_df.groupby(['client_number', 'country_codes', 'currency_code', 'status_group'])['amount'].sum().reset_index()
pivot_df = aggregated_df.pivot(index=['client_number', 'country_codes', 'currency_code'], columns='status_group', values='amount').reset_index()
pivot_df = pivot_df[['client_number', 'country_codes', 'currency_code', 'completed']]
pivot_df.rename(columns={'client_number': 'client_number', 'completed': 'total_transactions_completed'}, inplace=True)
pivot_df.fillna(0, inplace=True)
clients_completed_payment = pivot_df.sort_values(by=['client_number', 'total_transactions_completed', 'country_codes', 'currency_code'], ascending=[True, False, True, True]).reset_index(drop=True)
clients_completed_payment
```

```python
total_transactions_by_currency = clients_completed_payment.groupby(['client_number', 'currency_code'])['total_transactions_completed'].sum().reset_index()
def calculate_total_transactions(row):
    formatted_amount = "{:.2f}".format(row['total_transactions_completed'])
    return f"Total transactions completed for customer {row['client_number']} are {row['currency_code']} {formatted_amount}"
total_transactions_by_currency['summary'] = total_transactions_by_currency.apply(calculate_total_transactions, axis=1)
sorted_summaries = total_transactions_by_currency.sort_values(by='total_transactions_completed', ascending=False)
for summary in sorted_summaries['summary']:
    print(summary)
```

```python
eur_transactions = clients_completed_payment[clients_completed_payment['currency_code'] == 'EUR']
total_transactions_eur = eur_transactions.groupby('client_number')['total_transactions_completed'].sum()
plt.figure(figsize=(10, 6))
wedges, texts, autotexts = plt.pie(total_transactions_eur, autopct=lambda p: f'{p:.2f}%', textprops={'fontweight': 'bold'},
                                   startangle=140, counterclock=False, wedgeprops={'edgecolor': 'gray'})
legend_labels = [f'{client} - {pct:.2f}%' for client, pct in zip(total_transactions_eur.index, total_transactions_eur / total_transactions_eur.sum() * 100)]
plt.legend(wedges, legend_labels, title="Client Numbers and %", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(texts, size=8)
plt.setp(autotexts, size=10, fontweight='bold')
plt.title('Total Transactions Completed by Customer (Currency: EUR)')
plt.show()
```

## Failed payments

Similar to our previous approach, we will analyze transactions per customer and country, but this time with a focus on the failed transactions.

```python
failed_payments_df = payments_df[payments_df['status_group'] == 'failed']
aggregated_failed_df = failed_payments_df.groupby(['client_number', 'country_codes', 'currency_code', 'status_group'])['amount'].sum().reset_index()
pivot_failed_df = aggregated_failed_df.pivot(index=['client_number', 'country_codes', 'currency_code'], columns='status_group', values='amount').reset_index()
pivot_failed_df = pivot_failed_df[['client_number', 'country_codes', 'currency_code', 'failed']]
pivot_failed_df.rename(columns={'client_number': 'client_number', 'failed': 'total_transactions_failed'}, inplace=True)
pivot_failed_df.fillna(0, inplace=True)
clients_failed_payment = pivot_failed_df.sort_values(by=['client_number', 'total_transactions_failed', 'country_codes', 'currency_code'], ascending=[True, False, True, True]).reset_index(drop=True)
clients_failed_payment
```

```python
total_transactions_by_currency_failed = clients_failed_payment.groupby(['client_number', 'currency_code'])['total_transactions_failed'].sum().reset_index()
def calculate_total_transactions_failed(row):
    formatted_amount = "{:.2f}".format(row['total_transactions_failed'])
    return f"Total transactions failed for customer {row['client_number']} in {row['currency_code']} are {formatted_amount}"
total_transactions_by_currency_failed['summary_failed'] = total_transactions_by_currency_failed.apply(calculate_total_transactions_failed, axis=1)
sorted_summaries = total_transactions_by_currency_failed.sort_values(by='total_transactions_failed', ascending=False)
for summary_failed in sorted_summaries['summary_failed']:
    print(summary_failed)
```

```python
eur_transactions_failed = clients_failed_payment[clients_failed_payment['currency_code'] == 'EUR']
total_transactions_eur_failed = eur_transactions_failed.groupby('client_number')['total_transactions_failed'].sum()
plt.figure(figsize=(10, 6))
wedges, texts, autotexts = plt.pie(total_transactions_eur_failed, autopct=lambda p: f'{p:.2f}%', textprops={'fontweight': 'bold'},
                                   startangle=140, counterclock=False, wedgeprops={'edgecolor': 'gray'})
legend_labels = [f'{client} - {pct:.2f}%' for client, pct in zip(total_transactions_eur_failed.index, total_transactions_eur_failed / total_transactions_eur_failed.sum() * 100)]
plt.legend(wedges, legend_labels, title="Client Numbers and %", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
plt.setp(texts, size=8)
plt.setp(autotexts, size=10, fontweight='bold')
plt.title('Total Transactions Failed by Customer (Currency: EUR)')
plt.show()
```

## Combined Completed and Failed payments

```python
total_transactions_by_currency = clients_completed_payment.groupby(['client_number', 'currency_code'])[['total_transactions_completed']].sum()

total_transactions_by_currency_failed = clients_failed_payment.groupby(['client_number', 'currency_code'])[['total_transactions_failed']].sum()

merged_data = pd.merge(total_transactions_by_currency, total_transactions_by_currency_failed, how='outer', left_index=True, right_index=True)
merged_data = merged_data.fillna(0)

merged_data['completed_percentage'] = (merged_data['total_transactions_completed'] / (merged_data['total_transactions_completed'] + merged_data['total_transactions_failed'])) * 100
merged_data['failed_percentage'] = (merged_data['total_transactions_failed'] / (merged_data['total_transactions_completed'] + merged_data['total_transactions_failed'])) * 100

for index, row in merged_data.iterrows():
    print(f"Customer {index[0]} - Currency: {index[1]} | Completed: {row['completed_percentage']:.2f}% | Failed: {row['failed_percentage']:.2f}%")
```

```python
total_transactions_by_currency = clients_completed_payment.groupby(['currency_code'])[['total_transactions_completed']].sum()
total_transactions_by_currency_failed = clients_failed_payment.groupby(['currency_code'])[['total_transactions_failed']].sum()
merged_data = pd.merge(total_transactions_by_currency, total_transactions_by_currency_failed, how='outer', left_index=True, right_index=True)
merged_data = merged_data.fillna(0)  # Fill NaN values with 0
merged_data['completed_percentage'] = (merged_data['total_transactions_completed'] / (merged_data['total_transactions_completed'] + merged_data['total_transactions_failed'])) * 100
merged_data['failed_percentage'] = (merged_data['total_transactions_failed'] / (merged_data['total_transactions_completed'] + merged_data['total_transactions_failed'])) * 100
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
eur_data = merged_data.loc['EUR']
eur_data[['completed_percentage', 'failed_percentage']].plot(kind='bar', ax=axes[0], color=['green', 'red'])
axes[0].set_title('EUR Transactions')
axes[0].set_ylabel('Percentage')
czk_data = merged_data.loc['CZK']
czk_data[['completed_percentage', 'failed_percentage']].plot(kind='bar', ax=axes[1], color=['blue', 'orange'])
axes[1].set_title('CZK Transactions')
axes[1].set_ylabel('Percentage')
plt.tight_layout()
plt.show()
```

From our observations, it's evident that completed transactions make up nearly 80% of the total transactions in both currencies.

## Days

### Completed transactions

Let's delve deeper into the transactions on a daily basis.

```python
payment_statuses_df['created_at'] = pd.to_datetime(payment_statuses_df['created_at'])
completed_transactions_df = payment_statuses_df[payment_statuses_df['payments_status_group'] == 'completed'].copy()
completed_transactions_df['date_completed'] = completed_transactions_df['created_at'].dt.date
transactions_per_day_completed = completed_transactions_df.groupby('date_completed').size()
plt.figure(figsize=(20, 10))
transactions_per_day_completed.plot(kind='line', marker='o')
plt.title('Completed Transactions per Day')
plt.ylabel('Number of Completed Transactions')
unique_dates_completed = completed_transactions_df['date_completed'].unique()
day_date_labels_completed = [f'{date.strftime("%A")} {date}' for date in unique_dates_completed]
for date in unique_dates_completed:
    week_start = date - pd.Timedelta(days=date.weekday())
    week_end = week_start + pd.Timedelta(days=6)
    week_transactions = transactions_per_day_completed[
        (transactions_per_day_completed.index >= week_start) & 
        (transactions_per_day_completed.index <= week_end)
    ]    
    if not week_transactions.empty:
        max_transactions_day = week_transactions.idxmax()
        if date == max_transactions_day:
            plt.axvline(x=date, color='red', linestyle='--')
plt.xticks(unique_dates_completed, day_date_labels_completed, rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The graph displays all completed transactions for each day in our dataset. Additionally, red lines are visible on the graph, indicating the days of the week with the highest completed transaction amounts.

## Speed of transactions

### Completed

When discussing payments, transaction speed holds paramount importance. Therefore, we will analyze the time it takes for the system to complete a transaction.

```python
completed_payments_df = payment_statuses_df[payment_statuses_df['payments_status_group'] == 'completed']
started_payments_df = payment_statuses_df[payment_statuses_df['payments_status_group'] == 'started']
merged_payments_df = pd.merge(started_payments_df, completed_payments_df, on='payment_number', suffixes=('_started', '_completed'))
merged_payments_df['created_at_started'] = pd.to_datetime(merged_payments_df['created_at_started'])
merged_payments_df['updated_at_completed'] = pd.to_datetime(merged_payments_df['updated_at_completed'])
merged_payments_df['processing_time'] = merged_payments_df['updated_at_completed'] - merged_payments_df['created_at_started']
merged_payments_df['days'] = merged_payments_df['processing_time'].dt.days
merged_payments_df['hours'] = merged_payments_df['processing_time'].dt.components['hours']
merged_payments_df['minutes'] = merged_payments_df['processing_time'].dt.components['minutes']
merged_payments_df['seconds'] = merged_payments_df['processing_time'].dt.components['seconds']
time_transactions = pd.merge(merged_payments_df, payments_df[['payment_number', 'country_codes']], on='payment_number')
time_transactions = time_transactions[['payment_number', 'days', 'hours', 'minutes', 'seconds', 'country_codes']]
time_transactions
```

## Daily Transaction Peaks

### Completed

```python
completed_transactions_df = payments_df[payments_df['status_group'] == 'completed'].copy()
completed_transactions_df['transaction_time'] = completed_transactions_df['created_at'].dt.time
completed_transactions_hour = completed_transactions_df[['payment_number', 'country_codes', 'transaction_time']]
completed_transactions_hour
```

```python
unique_countries = completed_transactions_hour['country_codes'].unique()
hour_ranges = pd.date_range(start='00:00:00', end='23:59:59', freq='H').time
for country_code in unique_countries:
    country_data = completed_transactions_hour[completed_transactions_hour['country_codes'] == country_code].copy()
    country_data['transaction_hour'] = country_data['transaction_time'].apply(lambda x: x.hour)
    grouped = country_data.groupby(pd.cut(country_data['transaction_hour'], bins=range(0, 25)))
    transaction_counts = grouped.size()
    plt.figure(figsize=(10, 6))
    transaction_counts.plot(kind='bar', rot=0)
    plt.xticks(range(24), hour_ranges, rotation=90)
    plt.xlabel('Hour Range')
    plt.ylabel('Number of Transactions')
    plt.title(f'Completed Transactions for Country {country_code}')
    plt.tight_layout()
    plt.show()
```

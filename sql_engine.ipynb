```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
payments_df = pd.read_excel("/Users/pablopavon/Documents/JOB_2023/Kevin/Homework_Task_Final.xlsx", sheet_name="Payments")
payment_statuses_df = pd.read_excel("/Users/pablopavon/Documents/JOB_2023/Kevin/Homework_Task_Final.xlsx", sheet_name="Payment Statuses")
```

```python
payments_df.columns = payments_df.columns.str.strip().str.replace(' ', '_').str.lower()
payments_df.rename(columns={'status_group_group': 'status_group'}, inplace=True)
payments_df['created_at'] = pd.to_datetime(payments_df['created_at'], format='%d/%m/%Y %H:%M:%S')
payments_df['updated_at'] = pd.to_datetime(payments_df['updated_at'], format='%d/%m/%Y %H:%M:%S')
payments_df['amount'] = payments_df['amount'].round(2)
payments_df
```

```python
payments_df.info()
```

```python
payments_df['created_at_date'] = payments_df['created_at'].dt.date
payments_df['created_at_hour'] = payments_df['created_at'].dt.time
payments_df['updated_at_date'] = payments_df['updated_at'].dt.date
payments_df['updated_at_hour'] = payments_df['updated_at'].dt.time

# Reordering the columns as requested
desired_order = [
    'payment_number', 'client_number', 'bank', 'country_codes',
    'created_at_date', 'created_at_hour', 'updated_at_date', 'updated_at_hour',
    'status_group', 'currency_code', 'amount'
]
payments_df = payments_df[desired_order]

# Convert the date columns to datetime64 data type
payments_df['created_at_date'] = pd.to_datetime(payments_df['created_at_date'])
payments_df['updated_at_date'] = pd.to_datetime(payments_df['updated_at_date'])

# Clean column names and round 'amount' column
payments_df.columns = payments_df.columns.str.strip().str.replace(' ', '_').str.lower()
payments_df['amount'] = payments_df['amount'].round(2)

payments_df
```

```python
payments_df.info()
```

```python
payment_statuses_df.columns = payment_statuses_df.columns.str.strip().str.replace(' ', '_').str.lower()
payment_statuses_df['created_at'] = pd.to_datetime(payment_statuses_df['created_at'], format='%d/%m/%Y %H:%M:%S')
payment_statuses_df['updated_at'] = pd.to_datetime(payment_statuses_df['updated_at'], format='%d/%m/%Y %H:%M:%S')


payment_statuses_df
```

```python
payment_statuses_df['created_at_date'] = payment_statuses_df['created_at'].dt.date
payment_statuses_df['created_at_hour'] = payment_statuses_df['created_at'].dt.time
payment_statuses_df['updated_at_date'] = payment_statuses_df['updated_at'].dt.date
payment_statuses_df['updated_at_hour'] = payment_statuses_df['updated_at'].dt.time
desired_order = [
    'payment_number', 'created_at_date', 'created_at_hour',
    'updated_at_date', 'updated_at_hour', 'payment_status', 'payments_status_group']
payment_statuses_df = payment_statuses_df[desired_order]
payment_statuses_df['created_at_date'] = pd.to_datetime(payment_statuses_df['created_at_date'])
payment_statuses_df['updated_at_date'] = pd.to_datetime(payment_statuses_df['updated_at_date'])
payment_statuses_df
```
